```python
for x in range(0, 203):
    hovno = 3**(10*x + 4) + 2**(10*x + 3) + 10
    print(f"(3^{10*x + 4} + 2^{10*x + 3} + 10) % 11 == 0: {hovno % 11 == 0}")
```

```python
for i in range(100,1000):
    x = i**2 % 1000
    y = (i+32)**2 % 1000
    if x == y:
        print(f'{i}: x: {x} y: {y} => {x == y}')
```
